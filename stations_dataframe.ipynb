```python
import geopy
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np
```

```python
def get_zipcode(df, geolocator, lat_field, lon_field):
    location = geolocator.reverse((df[lat_field], df[lon_field]))
    try:
        return location.raw['address']['postcode']
    except KeyError:
        return np.nan


geolocator = geopy.Nominatim(user_agent='ben')
```

```python
stations_df = pd.read_csv('http://web.mta.info/developers/data/nyct/subway/Stations.csv',
                         index_col = 'Station ID')
```

I cannot be certain, but I am fairly sure that there is a problem being generated by the length of some of the latitude and longitude values. Therefore, I am rounding them to 6 decimal places, which appears to be the value displayed by default:

```python
stations_df = stations_df.round({'GTFS Longitude': 6, 'GTFS Latitude': 6})
```

```python

stations_df['zipcode'] = stations_df.apply(get_zipcode, axis=1, geolocator=geolocator, lat_field='GTFS Latitude', lon_field='GTFS Longitude')
```
